### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda4.info()
```

De las 12 columnas entregadas con información solo 5 de estas nos sirven para el análisis solicitado. No se encuentran datos nulos en ninguna de las tiendas

#Manipulacion de datos

##Eliminación de celdas no necesarias

```python
tienda_depurada = tienda.drop(["Fecha de Compra", "Vendedor", "Lugar de Compra", "Método de pago", "Cantidad de cuotas"], axis=1)
tienda_depurada.info()
```

```python
tienda2_depurada = tienda2.drop(["Fecha de Compra", "Vendedor", "Lugar de Compra", "Método de pago", "Cantidad de cuotas"], axis=1)
tienda2_depurada.info()
```

```python
tienda3_depurada = tienda3.drop(["Fecha de Compra", "Vendedor", "Lugar de Compra", "Método de pago", "Cantidad de cuotas"], axis=1)
tienda3_depurada.info()
```

```python
tienda4_depurada = tienda4.drop(["Fecha de Compra", "Vendedor", "Lugar de Compra", "Método de pago", "Cantidad de cuotas"], axis=1)
tienda4_depurada.info()
```

#1. Análisis de facturación



##Ingresos totales de cada tienda

```python
ingresos_tienda = tienda_depurada["Precio"].sum()
print(ingresos_tienda)
```

```python
ingresos_tienda2 = tienda2_depurada["Precio"].sum()
print(ingresos_tienda2)
```

```python
ingresos_tienda3 = tienda3_depurada["Precio"].sum()
print(ingresos_tienda3)
```

```python
ingresos_tienda4 = tienda4_depurada["Precio"].sum()
print(ingresos_tienda4)
```

En este análisis encontramos que la tienda con menores ingresos es la cuarta obteniendo ingresos de 1.038.375.700, en comparación con los mayores ingresos de la primera tienda que a obtenido ingresos por 1.150.880.400 .

# 2. Ventas por categoría

```python
clasificación_tienda = tienda_depurada["Categoría del Producto"].value_counts()
print(clasificación_tienda)
```

```python
clasificación_tienda2 = tienda2_depurada["Categoría del Producto"].value_counts()
print(clasificación_tienda2)
```

```python
clasificación_tienda3 = tienda3_depurada["Categoría del Producto"].value_counts()
print(clasificación_tienda3)
```

```python
clasificación_tienda4 = tienda4_depurada["Categoría del Producto"].value_counts()
print(clasificación_tienda4)
```

En este análisis nos encontramos que todas las tiendas presentan valores similares, siendo los 3 articulos más vendidos los Muebles, Electronicos y juguestes, mientras que el menos vendido es Articulos del hogar, salvo en la cuarta tienda en donde los menos vendido son los Instrumentos musicales.

En el siguiente análisis nos encontramos con que la tienda con mejor promedio de calificaciones por parte de los clientes es la tienda 3, mientras que la que tiene el promedio más bajo es la tienda 1.

# 4. Productos más y menos vendidos

```python
producto_mas_vendido_tienda = tienda_depurada["Producto"].value_counts().head(1)
producto_menos_vendido_tienda = tienda_depurada["Producto"].value_counts().tail(1)
print(producto_mas_vendido_tienda)
print(producto_menos_vendido_tienda)
```

```python
producto_mas_vendido_tienda2 = tienda2_depurada["Producto"].value_counts().head(1)
producto_menos_vendido_tienda2 = tienda2_depurada["Producto"].value_counts().tail(1)
print(producto_mas_vendido_tienda2)
print(producto_menos_vendido_tienda2)
```

```python
producto_mas_vendido_tienda3 = tienda3_depurada["Producto"].value_counts().head(1)
producto_menos_vendido_tienda3 = tienda3_depurada["Producto"].value_counts().tail(1)
print(producto_mas_vendido_tienda3)
print(producto_menos_vendido_tienda3)
```

```python
producto_mas_vendido_tienda4 = tienda4_depurada["Producto"].value_counts().head(1)
producto_menos_vendido_tienda4 = tienda4_depurada["Producto"].value_counts().tail(1)
print(producto_mas_vendido_tienda4)
print(producto_menos_vendido_tienda4)
```

Segun los datos arrojados los articulos más vendidos son muy diferentes entre si: en la tienda 1 fueron microondas, en la tienda 2 fue un producto llamado "Iniciando con programación", en la tienda 3 fue un Kit de bancas y en la tienda 4 una cama box. En cambio, los productos menos vendidos fueron en la tienda 1 un celular ABXY, en la tienda 2 fueron juegos de mesa, en la tienda 3 fueron bloques de construccion y en la tienda 4 fueron guitarras electricas. Todo esto solo demuestra que cada tienda tiene tipos muy diferentes de clientes y con diferentes prioridades.
